```python
# Importing the random and Tk interfacemodules
import tkinter
import random


colours = ['Red','Blue','Green','Pink','Black',
		'Yellow','Orange','White','Purple','Brown'] # we put down the list of possible colour.
score = 0


timeleft = 60 # this is the  game time left 60 seconds.


def startGame(event): # we create a function that will start the game.
	
	if timeleft == 60:
		
		
		countdown() # we canstart the countdown timer.
		

	nextColour() #choosing the next colour


def nextColour(): #this is the function that we choose and select the next colour

	
	global score
	global timeleft

	
	if timeleft > 0:# if a game is currently playing

		
		e.focus_set() #we make the text entry box active.

		
		if e.get().lower() == colours[1].lower(): # if the colour typed by the player is equal to the colour of the text in the box
			
			score += 1

		
		e.delete(0, tkinter.END) # then we clear the text entry box.
		
		random.shuffle(colours)
		
		
		label.config(fg = str(colours[1]), text = str(colours[0])) #we are changing the colour to type in to a random colour value
		
		scoreLabel.config(text = "Score: " + str(score)) #we are changing the score everytime the player gets a point



def countdown(): #we are defining the countdown timer function

	global timeleft

	
	if timeleft > 0: # if a game is currently playing

		
		timeleft -= 1 #reduce the timer
		
		
		timeLabel.config(text = "Time left: "
							+ str(timeleft)) #updating the time left
								
		
		timeLabel.after(1000, countdown) #we are running the function again after 1 second





root = tkinter.Tk() # we are creating a GUI window


root.title("The colour game") #the title of the window


root.geometry("500x350") #we are selecting the size for the dimensions of the window


instructions = tkinter.Label(root, text = "Type in the colour"
						" of the words, and not the word text!",
									font = ('Arial', 14)) #the instructions to the label
instructions.pack()


scoreLabel = tkinter.Label(root, text = "Press enter to start",
									font = ('Arial', 12)) # we are adding a score label
scoreLabel.pack()


timeLabel = tkinter.Label(root, text = "Time left: " +
			str(timeleft), font = ('Arial', 12)) #adding the time labe;
			
timeLabel.pack()


label = tkinter.Label(root, font = ('Arial', 60)) #the label to display the colurs
label.pack()


e = tkinter.Entry(root) #entry box for the player to type in the colour


root.bind('<Return>', startGame) #start the game as soon as the player presses the key
e.pack()


e.focus_set() #set a focus entry box to display a window with an active entry widget


root.mainloop() #start the GUI
```
